### Missing data

- Important questions when thinking about missing data:


- How prevalent is the missing data? Is missing data random or does it have a pattern?


The answer to these questions is important for practical reasons because missing data can imply a reduction of the sample size. This can prevent us from proceeding with the analysis. Moreover, from a substantive perspective, we need to ensure that the missing data process is not biased and hidding an inconvenient truth.

```python
#checking the missing data
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

#### Observations:

- From the above output, We can see that the below list of columns have missing values,
 
total cases, new_deaths, new deaths_smoothed, new_cases_smoothed_per_million, new_deaths_per_million, median_age, aged_65_older,aged_70_older, gdp_per_capita, cardiovasc_death_rate, diabetes_prevalence, handwashing_facilities, life_expectancy, human_development_index.

- However, the percent of missing values in these columns are relatively very low when compared to the dataset in    whole. 


- Since it is negligible, We can drop the rows that have null values or we can impute them with Means or most frequent values in that columns

# Data Visualizations

## 1. HISTOGRAM

#### Description:

- A histogram is used to summarize discrete or continuous data. 


- It provides a visual representation of data distribution. 


- Histograms can display a large amount of data and the frequency of the data values. The median and distribution of the data can be determined by a histogram. In addition, it can show any outliers or gaps in the data.

```python
# Plotting the histogram for extreme_poverty and gdp_per_capita to check how the distribution is present
fig = plt.figure()

sns.displot(df["extreme_poverty"],height=4,aspect=2)

sns.displot(df["gdp_per_capita"],height=4,aspect=2)
# Showing the figure
plt.show()
```

#### Observations:

- People with lower income were affected the most 


- The Dataset is biased towards people with lower GDP per capita 


- Africa has the highest extreme poverty ratio as per the dataset


- Agenda to use histogram on the above data is because, We had to represent data with large frequencey which was         continous and widely distributed.


- We conclude by saying that the exact new case/death per million may not have been registered accurately for those     countries with extreme poverty  

## 2. BOX PLOT

#### Description:

- A boxplot is a standardized way of displaying the dataset based on a five-number summary: the minimum, the maximum,   the sample median, and the first and third quartiles.


- It can tell us about the data outliers and what their values are. 


- It can also tell us if your data is symmetrical, how tightly the data is grouped, and if and how your data is skewed.

```python
# Creating a box plot with subgroups
var1 = ['India','Pakistan','El Salvador','Moldova','Bosnia and Herzegovina','Dominican Republic','Algeria']
mask2=df['location'].isin(var1)
df2=df[mask2]

sns.catplot(x="location",y="stringency_index",
            data=df2,
            kind="box",aspect=2)
plt.xticks(rotation=90)
# Show plot
plt.show()
```

#### Observations:

- This boxplot describes the statistical summary of countries and their stringency_index.


- Bosnia was impacted the most with new case and new deaths/million since they had the lowest range of stringency       index (Stringency effects and impacts are captured in the next few slides) 


- While India & El Salvador have the highest stringency index but have been flexible, Algeria has been stringent         throughout.   


- Agenda for using boxplot for the above data is to understand the numerical data clearly and check the outliers

## 3. BAR CHART

#### Description:


- A bar chart is a graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. 


- A bar graph shows comparisons among discrete categories


- Primary difference between a histogram and a bar plot is that histogram works with quantitative data and bar plot works with categorical data. While histograms are used to show distributions of variables, Bar charts are used to compare variables. 

```python
plt.figure(figsize=(20,5))
sns.set(font_scale=2)
clrs = ['orange' if (x < max(df['total_deaths'])) else 'green' for x in df['total_deaths'] ]
# Creating a box plot with subgroups and omit the outliers
g = sns.barplot(x= 'location',y="total_deaths", ci = None, palette=clrs,
            data=df)
g.set_yscale("log")
plt.xticks(rotation=90)

# Show plot
plt.show()
```

#### Observations:

- India & Mexico have the highest number of death counts while Laos & Timor have the lowest number of death counts


- India and Mexico have the highest death rates, does it mean that these two countries also have the highest number of cases/million ? which is captured in the next few visualization


- We used barplot for the above data because, we are making the comparisions between categorical variables.


## 4. SCATTER PLOT

#### Description:

- Primarily used to observe and show relationships between two numeric variables.


- The dots in a scatter plot not only report the values of individual data points, but also patterns when the data are   taken as a whole

```python
#creating a scatter plot between new cases and new deaths per million.
plt.figure(figsize=(10,10))
sns.set(font_scale=1)
var = ['India','Moldova','Mexico','Bosnia and Herzegovina']
mask1=df['location'].isin(var)
df1=df[mask1]

sns.scatterplot( data=df1,x="new_deaths_per_million", y="new_cases_per_million", hue="location")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```

#### Observations:


- Bosnia experienced the most surge in cases and deaths/million


- Even though there were few new cases registered as compared to other countries their death rate was still on the       higher side


- Moldova reported close to 590 new cases/million, however, the death rate never crossed 12/million


- In Mexico, the death rate was high in spite of fewer new cases being registered


- Reason to pick scatter plot for the above is because it is the best way to represent the categorical variable "location" by comparing two different numerical variables

## 5. Line Chart

#### Description:

- It is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments.
 
 
- It is similar to a scatter plot except that the measurement points are ordered and joined with straight line segments. 


- A line chart is often used to visualize a trend in data over intervals of time â€“ a time series â€“ thus the line is often drawn chronologically. In these cases they are known as run charts.

```python
#creating a line chart to see rise in new cases over the past 2 years
sns.set(font_scale=1)
ax=sns.relplot(x="month", y="new_cases", hue = 'year',
            data=df, kind="line", 
            ci=None,
            markers=True,dashes=False, palette = 'Set1')
ax.fig.set_figwidth(15)
# Show plot
plt.xticks(rotation=90)
plt.show(ax)
```

#### Observations:

- In 2020, most new cases were registered in the month of September 


- In 2021, most new cases were registered in the month of May 


- New cases registered are more in 2021 when compared to 2020


- Agenda for using a line chart for above data is, because the data used is time series.

## 6. PIE CHART

#### Description:

- This is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. 


- It contains both numerical and categorical variables.


- In a pie chart, the arc length of each slice, is proportional to the quantity it represents.

## 7. HEATMAP

#### Description:

- It is a graphical representation of data that uses a system of color-coding to represent different values. 


- Heatmaps are used in various forms of analytics but are most commonly used to show user behaviour on specific of multiple variables with respect a unique feature.

```python
heatmap_data = pd.pivot_table(df1, values='total_cases', index=['month'], columns='year')
```

```python
sns.heatmap(heatmap_data,cmap="coolwarm")
plt.xlabel("years", size=10)
plt.ylabel("Months", size=10)
plt.title(" Total Cases Monthly", size=14)
plt.tight_layout()
```

#### Observations:

- This map describes the total cases with respect to month for year 2020 and 2021. While colour bad describes the frequency of the deaths registered in a month.


- May & June in 2021 has the highest number of cases recorded


- March & April in 2020 has the lowest number of cases recorded


- Total number of cases increased rapidly from March 2020 through June 2021 


- Motive to use this type of graph is to provide more visual approach in understanding the numerical variables and give an instant overview of their performance.
